```python
from IPython import display
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('seaborn-colorblind')

heading = np.arange(0, 2 * np.pi, 0.02 * np.pi)

x = np.random.rand(100)

y = np.random.rand(100)

for k in heading:
    plt.clf()
    y = np.where(y < 1, y + np.sin(k/1000), y - 1)
    x = np.where(x < 1, x + np.sin(k/1000), x -1)
    plt.scatter(x, y)
    plt.axis([0,1,0,1])
    display.display(plt.gcf())
    display.clear_output(wait=True)
```
