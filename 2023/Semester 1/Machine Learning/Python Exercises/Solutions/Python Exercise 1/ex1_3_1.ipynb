```python
from IPython import display
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')

# Generate the heading array
heading = np.cos(np.arange(0, 2 * np.pi, 0.02 * np.pi))

x = np.random.uniform(0, 1, 100)

y = np.random.uniform(0, 1, 100)

m = 0
while(m < 100):
    plt.clf()
    y = np.where(y < 1, y + heading/1000, y -1)
    x = np.where(x < 1, x + heading/1000, x -1)
    plt.scatter(x, y)
    plt.axis([0,1,0,1])
    display.display(plt.gcf())
    display.clear_output(wait=True)
    m += 1
    
```
